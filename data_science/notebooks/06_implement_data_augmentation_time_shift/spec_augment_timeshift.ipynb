```python
import matplotlib.pylab as plt 
import warnings
import pandas as pd
import os
import random

from librosa import display
from src.processing import preprocessing
from src.processing import data_augmentation
```

```python
random.seed(111)

os.environ["CUDA_VISIBLE_DEVICES"] = "0"
warnings.filterwarnings("ignore")
```

```python
data_df = pd.read_json('/datasets/UrbanSound8K/processed/mean_mfcc_data.json')
data_df = preprocessing.filter_mfccs(data_df)
```

```python
i = 1
print(data_df.label.iloc[i])
display.specshow(data_df.feature.iloc[i], x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.show()
```

```python
for i in range(10):
    print(data_df.label.iloc[i])
    display.specshow(data_df.feature.iloc[i], x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.show()
    display.specshow(data_augmentation.time_shift_mfcc(data_df.feature.iloc[i]), x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.show()
```
