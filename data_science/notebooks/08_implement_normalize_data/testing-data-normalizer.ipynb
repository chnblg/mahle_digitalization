```python
import pandas as pd
import matplotlib.pylab as plt 

from librosa import display
from src.processing import preprocessing
```

```python
data_df_test1 = pd.read_json('/datasets/UrbanSound8K/processed/mean_mfcc_data.json')
data_df_test1 = preprocessing.filter_mfccs(data_df_test1)

data_df_test2 = pd.read_json('/datasets/UrbanSound8K/processed/mean_mfcc_data.json')
data_df_test2 = preprocessing.filter_mfccs(data_df_test2)
```

```python
data_df_test2 = preprocessing.normalize_mfccs(data_df_test2)
```

```python
for i in range(50):
    print(i)
    display.specshow(data_df_test1.feature.iloc[i], x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.show()

    print(i)
    display.specshow(data_df_test2.feature.iloc[i], x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.show()
```
