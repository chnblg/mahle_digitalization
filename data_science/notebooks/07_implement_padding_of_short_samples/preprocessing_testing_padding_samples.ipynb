```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt 
import warnings
import os

from librosa import display

warnings.filterwarnings("ignore")
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
```

```python
data_df = pd.read_json('/datasets/UrbanSound8K/processed/mean_mfcc_data.json')
```

```python
for i in range(15):
    mfc_array = np.array(data_df['feature'].iloc[i])
    if mfc_array.shape[1] < 173:
        padding_factor = 173 - mfc_array.shape[1]
        print(i)
        print(mfc_array.shape)
        print(padding_factor)
        mfc_array = np.pad(mfc_array, [(0, 0), (0, padding_factor)], mode='constant', constant_values=(0))
        print(mfc_array.shape)
    data_df['feature'].iloc[i] = mfc_array
```

```python
for i in range(15):
    print(data_df.label.iloc[i])
    display.specshow(data_df.feature.iloc[i], x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.show()
```
